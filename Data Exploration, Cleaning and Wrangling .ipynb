```python
df2['CompFreq']=df2['CompFreq'].replace('Office', np.NaN)
df2['CompFreq']
```

```python
df2['CompTotal']=df2['CompTotal'].replace('Office', np.NaN)
df2['CompTotal']
```

```python
df2.filter(items=['CompFreq','CompTotal'])
```

```python
df2['CompFreq'].value_counts()
```

```python
annualcomp=[]
for x,y in zip(df2['CompFreq'], df2['CompTotal']):
    if x=='Monthly':
        annualcomp.append(y*12)
    elif x=='Weekly':
        annualcomp.append(y*52)
    else:
        annualcomp.append(y)

          
df2['NormalizedAnnualCompensation']=anncomp
df2[['NormalizedAnnualCompensation']]
```
